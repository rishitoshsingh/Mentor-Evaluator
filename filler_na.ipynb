```python
import numpy as np
from datetime import timedelta
```

```python
def _get_middle_values(start, end, n):
    diff = end - start
    req_diff = diff / (n+1)
    times = []
    for i in range(n):
        times.append(start + req_diff*(i+1))
    return times
```

```python
def _fill(series):
    gap_found = False
    start_i, end_i = None, None
    for i in range(len(series)):
        if not np.isnat(series[i]) :
            if gap_found == True:
                end_i = i
                values = _get_middle_values(series[start_i], series[end_i], (end_i-start_i-1))
                for temp_i, value in zip(range(start_i+1,end_i), values):
                    series[temp_i] = value
                print(series[start_i:end_i])
                gap_found = False
            else:
                continue
        else:
            if gap_found == False:
                gap_found = True
                start_i = i-1     
```
